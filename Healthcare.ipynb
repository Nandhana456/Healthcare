### **11.Correlation Between Age and Gender**

```python
sns.scatterplot(data=df,x='Age',y='Gender')
plt.title('correlation between age and gender')
plt.xlabel('Age',fontweight='bold')
plt.xlabel('Gender',fontweight='bold')
plt.show()

```

```python
sns.heatmap(df.corr(),vmin=-1,vmax=1)
plt.show
```

```python
sns.heatmap(df.corr(),vmin=-4,vmax=4)
plt.show
```

```python
corr = df.select_dtypes('number').corr()
sns.heatmap(data = corr, annot = True, linewidths = 1)
plt.title('Correlation between features', fontweight = 'bold')
plt.show()
```

# 12.**Histogram**

```python
df.Hospital.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of Doctor by Hospital")
plt.ylabel('Number of Doctor')
plt.xlabel('Hospital');
```

```python
df.Hospital.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of Name by Gender")
plt.ylabel('Number of Gender')
plt.xlabel('Name');
```

# **13.Heat Maps**
Heat Maps is a type of plot which is necessary when we need to find the dependent variables. One of the best way to find the relationship between the features can be done using heat maps.



```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
plt.figure(figsize=(20,5))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

# **Piechart**

```python
y=np.array([35,25,15,5])
mylabels=["Name","Age","Gender","Blood Type"]
plt.pie(y,labels=mylabels)
plt.show()
```

# **14.Scatterplot**

We generally use scatter plots to find the correlation between two variables. Here the scatter plots are plotted between Horsepower and Price and we can see the plot below. With the plot given below, we can easily draw a trend line. These features provide a good scattering of points.



```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['Age'], df['Billing Amount'])
ax.set_xlabel('Age')
ax.set_ylabel('Billing Amount')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['Name'], df['Blood Type'])
ax.set_xlabel('Name')
ax.set_ylabel('Blood Type')
plt.show()
```
